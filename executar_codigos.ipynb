```python
import pandas as pd
import random
from IPython.display import clear_output

watchlist_pd = pd.read_csv('watchlist.csv')

watchlist = watchlist_pd.loc[watchlist_pd["Title Type"] == 'Movie',"Title"].to_list()

watchlist_size = len(watchlist)

while watchlist_size > 1:
    movies =  random.sample(watchlist, 2)
    print(f"Choose a movie to watch:\n1. {movies[0]}\n2. {movies[1]}\n")

    choose = input("Select a movie to watch (1 or 2) or 'q' to quit:\n")
    clear_output(wait=False)

    if choose == '1':
        watchlist.remove(movies[1])
    elif choose == '2':
        watchlist.remove(movies[0])
    elif choose.lower() == 'q':
        print("Exiting the movie chooser. Enjoy your movie!")
        break
    
    watchlist_size = len(watchlist)
    print(f"{watchlist_size} movies left in the watchlist.\n")

print(f"Your selected movie to watch is: {watchlist[0]}")
```
